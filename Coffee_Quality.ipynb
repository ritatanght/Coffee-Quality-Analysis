# About Dataset
## Coffee Quality Institute
The Coffee Quality Institute (CQI) is a non-profit organization that works to improve the quality and value of coffee worldwide. It was founded in 1996 and has its headquarters in California, USA.

CQI's mission is to promote coffee quality through a range of activities that include research, training, and certification programs. The organization works with coffee growers, processors, roasters, and other stakeholders to improve coffee quality standards, promote sustainability, and support the development of the specialty coffee industry.

## Data
CQI maintains a web database that serves as a resource for coffee professionals and enthusiasts who are interested in learning about coffee quality and sustainability. The database includes a range of information on coffee production, processing, and sensory evaluation. It also contains data on coffee genetics, soil types, and other factors that can affect coffee quality.

## Sensory evaluations (coffee quality scores)
- Aroma: Refers to the scent or fragrance of the coffee.
- Flavor: The flavor of coffee is evaluated based on the taste, including any sweetness, bitterness, acidity, and other flavor notes.
- Aftertaste: Refers to the lingering taste that remains in the mouth after swallowing the coffee.
- Acidity: Acidity in coffee refers to the brightness or liveliness of the taste.
- Body: The body of coffee refers to the thickness or viscosity of the coffee in the mouth.
- Balance: Balance refers to how well the different flavor components of the coffee work together.
- Uniformity: Uniformity refers to the consistency of the coffee from cup to cup.
- Clean Cup: A clean cup refers to a coffee that is free of any off-flavors or defects, such as sourness, mustiness, or staleness.
- Sweetness: It can be described as caramel-like, fruity, or floral, and is a desirable quality in coffee.

**PLEASE NOTE: 'Total Cup Points' is literally the total of 10 features given above.**

## Defects
Defects are undesirable qualities that can occur in coffee beans during processing or storage. Defects can be categorized into two categories: Category One and Category Two defects.

- *Category One defects* are primary defects that can be perceived through visual inspection of the coffee beans. These defects include Black beans, sour beans, insect-damaged beans, fungus-damaged beans, etc.

- *Category Two defects* are secondary defects that are more subtle and can only be detected through tasting. These defects include Over-fermentation, staleness, rancidness, chemical taste, etc.

```python
# Make the values in Variety to be consistent and correct for typos
df['Variety'] = df['Variety'].fillna('unknown')
df['Variety'] = df['Variety'].replace('unknow', 'unknown')
df['Variety'].value_counts()
```

```python
# Copy the values from the Variety column to the Variety_Group column and perform groupings
# First, convert the values into lowercase for easier comparison
df['Variety_Group'] = df.Variety.str.lower()
# Group related variety names for consistencies
df['Variety_Group'] = df['Variety_Group'].replace({
                                                  'bourbon sidra': 'bourbon',
                                                  'yellow bourbon': 'bourbon',
                                                  'red bourbon': 'bourbon',
                                                  'yellow catuai': 'catuai',
                                                  'castillo paraguaycito': 'castillo'
                                                   })

# Rows with multiple varieties will be classified as 'Blend'
# All other varieties appearing only once in the dataset will be grouped into 'Others'
variety_list = df['Variety_Group'].value_counts()
variety_list = variety_list[variety_list > 1].index.tolist()
variety_list

# import relevant library to perform split by mulitple delimiters
import re
delimiters_pattern = r',|\s|&|-|\s+and\s+|\+'

def count_variety(variety):
  if variety in variety_list:
    return variety
  else:
    # this could be a blend or other
    parts = re.split(delimiters_pattern, variety)
    parts = [part.strip() for part in parts if part.strip()]
    if len(parts) > 1:
      return 'Blend'
    else:
      return 'Others'

df['Variety_Group'] = df['Variety_Group'].apply(count_variety)
df['Variety_Group'] = df['Variety_Group'].str.title()

df['Variety_Group'] = df['Variety_Group'].replace({
    # Replace 'Unknown' with 'Others'
    'Unknown': 'Others',
    # Make the SL and SHG varieties into proper cases as well
    'Sl34': 'SL34',
    'Sl14': 'SL14',
    'Sl28': 'SL28',
    'Shg': 'SHG',
})


df['Variety_Group'].value_counts()
```

The `Others` category groups `Unknown` varieties and those with a single instance in this dataset.

The `Blend` category, on the other hand, consists of entries representing multiple varieties, each appearing only once in this dataset.

```python
df['Processing Method'].value_counts()
# The 3 main board categories should be Washed, Natural, and Honey. The rest will be grouped into 'Others'
```

```python
# Create a function to perform grouping
def group_processing_method(method):
    if method.startswith('Washed'):
        return 'Washed'
    elif method.startswith('Natural'):
        return 'Natural'
    elif method.endswith('honey'):
        return 'Honey'
    else:
        return 'Others'

# Fill missing values with 'Others'
df['Processing Method'] = df['Processing Method'].fillna('Others')

# Create a new column for the categorized processing method group
df['Processing Method Category'] = df['Processing Method'].apply(group_processing_method)
df['Processing Method Category'].value_counts()
```

```python
# Check to ensure there are no missing values after cleaning
df.isnull().sum()
```

```python
df['Altitude'] = df['Altitude'].str.replace(r' - |~| A ', '-', regex=True)
df.Altitude.unique()
```

Since coffee grown at higher elevations tends to be of higher quality, I will categorize the `Altitude` values into groups for further analysis. While the unit for altitude is not explicitly provided,  it is assumed to be in **meters (m)** above sea level based on the context:
- 1500 or above --> Very High
- Between 1200 and 1500 --> High
- Between 900 and 1200 --> Medium
- 900 or below --> Low

It is also noted that unusually high numbers, such as '4895-5650' and '5400', are considered data entry errors. They are to be fixed...

```python
# Convert the range values into single number for further analysis
def get_altitude(altitude):
  # for fields with single number
  if isinstance(altitude, (int, float)) or altitude.isdigit():
     return int(altitude)
  # for fields with more than one number in the field, use the average of the 2 numbers
  else:
    nums = altitude.split('-')
    nums = [int(num) for num in nums]
    return np.mean(nums)

df['Altitude_Adjusted'] = df['Altitude'].apply(get_altitude)
print(df['Altitude_Adjusted'].describe())

# Draw a boxplot to visualize the distribution of Altitude after adjusted for range values
sns.boxplot(df['Altitude_Adjusted'], width=0.3)
plt.title('Distribution of Adjusted Altitude')
plt.ylabel('Altitude (meters)')
plt.xticks([])
# 4 outliers can be clearly identified from the boxplot below.
plt.annotate('4 outliers',
              (0.05,5000),
              bbox=dict(
                boxstyle="round,pad=0.5",
                facecolor='salmon',
                edgecolor='salmon',
                alpha=0.7
              ))
plt.show()
```

```python
# To investigate further, list the 4 outliers which have `Altitude_Adjusted` larger than 4000
altitude_outliers = df[df['Altitude_Adjusted'] > 4000]
altitude_outliers
# Subsequent research indicates these values are likely in feet.
```

```python
# Convert the 4 altitude values from feet to meters for consistency
df.loc[altitude_outliers.index, 'Altitude_Adjusted'] = df.loc[altitude_outliers.index, 'Altitude_Adjusted'] * 0.3048
df.loc[altitude_outliers.index, 'Altitude_Adjusted']
```

```python
# put the beans into groups based on the adjusted altitude values
def altitude_group(altitude):
    alt_val = get_altitude(altitude)
    if alt_val >= 1500:
      return 'Very High'
    elif alt_val >= 1200:
      return 'High'
    elif alt_val >= 900:
      return 'Medium'
    else:
      return 'Low'

df['Altitude_Category'] = df['Altitude_Adjusted'].apply(altitude_group)
df['Altitude_Category'].value_counts()
```

## Statistical Analysis

```python
from scipy import stats
```

### Is there a difference in `Total Cup Points` with difference in `Altitude_Adjusted`?

Using α = 0.05

Hypothesis:

*   H0: μ1 = μ2 = μ3 = μ4 (the four altitude groups' means are equal)
*   H1: At least one of the means differ


```python
very_high = df[df['Altitude_Category'] == 'Very High']['Total Cup Points']
high = df[df['Altitude_Category'] == 'High']['Total Cup Points']
medium = df[df['Altitude_Category'] == 'Medium']['Total Cup Points']
low = df[df['Altitude_Category'] == 'Low']['Total Cup Points']

stat, p_levene = stats.levene(very_high, high, medium, low)
print(f"Levene’s test p-value: {p_levene}")
# Since the p-value > 0.05, we assume the groups have equal variances. We will proceed with ANOVA.
```

```python
f_statistic, p_value = stats.f_oneway(very_high, high, medium, low)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

*Conclusion*: Since the p-value < 0.05, we reject the null hypothesis as there is significant evidence that at least one of the means differ.


---



### Is there a relationship between `Altitude_Adjusted` and `Total Cup Points`?
Using α = 0.05

Hypothesis:

*   H0: No relationship between Altitude and Total Cup Points
*   H1: There is a relationship between Altitude and Total Cup Points

```python
corr_coef, p_value = stats.pearsonr(df['Altitude_Adjusted'], df['Total Cup Points'])

print(f"Pearson correlation coefficient: {corr_coef:4f}, p-value: {p_value:4f}")
```

```python
sns.regplot(x='Altitude_Adjusted', y='Total Cup Points', data=df, line_kws={'color':'red'})
plt.title('Relationship between Altitude and Total Cup Points')
plt.xlabel('Altitude (meter)')
plt.ylabel('Total Cup Points')
plt.show()
```

*Conclusion*: Since the p-value < 0.05, we reject the null hypothesis and conclude that there is a statistically significant relationship between Altitude and Total Cup Points. The Pearson correlation coefficient of 0.1727 indicates a very weak positive relationship between the two variables. This suggests that coffee grown at *higher altitudes* tends to achieve *higher Total Cup Points*, although the relationship is weak.

---



### Is there a difference in `Flavor` with difference in Processing Method?

Using α = 0.05

Hypothesis:

*   H0: μ1 = μ2 = μ3 = μ4 (the four `Processing Method Category` means are equal)
*   H1: At least one of the means differ

```python
washed = df[df['Processing Method Category'] == 'Washed']['Flavor']
natural = df[df['Processing Method Category'] == 'Natural']['Flavor']
honey = df[df['Processing Method Category'] == 'Honey']['Flavor']
others = df[df['Processing Method Category'] == 'Others']['Flavor']

f_statistic, p_value = stats.f_oneway(washed, natural, honey, others)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

*Conclusion*: Since the p-value > 0.05, we fail to reject the null hypothesis. There is no statistically significant evidence to suggest that the mean Flavor score differs across the four processing methods.
